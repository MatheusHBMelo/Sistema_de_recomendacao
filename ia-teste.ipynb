```python
import pandas as pd 
import numpy as np

ratings = pd.read_csv('../input/reviews/olist_order_reviews_dataset.csv', index_col=0)
items = pd.read_csv('../input/orderitems/olist_order_items_dataset.csv', index_col=0)
products = pd.read_csv('../input/product/olist_products_dataset.csv', index_col=0)
orders = pd.read_csv('../input/orders/olist_orders_dataset.csv', index_col=0)
```

```python
# Dentro do dataset de pedidos, utilizaremos apenas duas colunas: ID do pedido e ID do cliente
orders.drop(orders.columns[[1,2,3,4,5,6]], axis=1, inplace=True)
orders.head()
```

```python
# Exibe o dataset dos produtos
products.drop(products.columns[[1,2,3,4,5,6,7]], axis=1, inplace=True)
products.head()
```

```python
# Dataset que contém as avaliações feitas pelos clientes.

ratings.drop(ratings.columns[[2,3,4]], axis=1, inplace=True)
ratings.head()
```

```python
# Dataset de pedido onde temos o ID do pedido relacionado com o ID do produto

items.drop(items.columns[[0,2,3,4,5]], axis=1, inplace=True)
items.head()
```

```python
# Merge com o dataset de Avaliações, fazendo o relacionamento mostrando quais foram os pedidos relacionados com os pedidos, exibindo o preço e score

items=items.merge(ratings, on='order_id')
items.head(5)
```

```python
# Merge com o dataset de pedido (orders), mostrando o ID do cliente que fez determinado pedido.
# Merge com o dataset do produto(products), mostrando os ID's do produtos e suas caracteristicas.

# O objetivo é mostrar qual cliente fez determinado pedido, mostrando o ID do produto.

items=items.merge(orders, on='order_id')
items=items.merge(products, on='product_id')
items.head(5)
```

```python
# Reordena as colunas para aplicar o 'surprise'
column_names = ["customer_id", "product_id", "review_score"]
items = items.reindex(columns=column_names)
items.head()
```

```python
# Importando pacotes
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise import Reader, SVD
import os
from surprise.model_selection import train_test_split
from sklearn.decomposition import TruncatedSVD
from surprise.model_selection import cross_validate
```

```python
# Lendo o dataset
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(items, reader=reader)

# Fazendo split dos dados / Test size de 30%
trainset, testset = train_test_split(data, test_size=0.3,random_state=10)

# Criação de um modelo baseado em item 
algo = KNNWithMeans(k=5, sim_options={'user_based': False})
algo.fit(trainset)

# Teste do modelo
test_pred = algo.test(testset)
```

```python
#Obtem o MAE
print("Item-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

```python
svd = SVD()
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
